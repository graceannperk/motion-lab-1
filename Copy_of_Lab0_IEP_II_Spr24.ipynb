````python
### TODO: YOUR CODE HERE
def computeTotalField(point_charges, observation_point):
  total_field = np.zeros(2)
  for charge in point_charges:
    field = charge.computeField(observation_point)
    total_field += np.array(field)
    return tuple(total_field)

def plotTotalField(point_charges, observation_point, xlim=(-10, 10), ylim=(-10, 10), scale = 10.0):
    x = np.linspace(xlim[0], xlim[1], 20)
    y = np.linspace(ylim[0], ylim[1], 20)
    X, Y = np.meshgrid(x, y)
    Ex, Ey = np.zeros_like(X), np.zeros_like(Y)
    for i in range(len(x)):
        for j in range(len(y)):
            field = computeTotalField(point_charges, (x[i], y[j]))
            Ex[j, i], Ey[j, i] = field
    magnitude = np.sqrt(Ex**2 + Ey**2)
    magnitude[magnitude == 0] = 1
    Ex /= magnitude
    Ey /= magnitude
    plt.quiver(X, Y, Ex, Ey, scale=20)
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('Total Electric Field (2D)')
    plt.xlim(xlim)
    plt.ylim(ylim)
    plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt

class pointCharge:
    def __init__(self, charge, position):
        self.charge = charge
        self.position = position

    def computeField(self, observation_point):
        r = np.array(observation_point)[:2] - np.array(self.position)[:2]
        r_mag = np.linalg.norm(r)
        if r_mag == 0:
            return np.zeros(2)  # If the observation point coincides with the charge position, return zero field
        field = (self.charge / (4 * np.pi * r_mag**3)) * r
        return field

def generate_surface_points(num_points):
    # Generate uniformly distributed points on the surface of a sphere
    phi = np.random.uniform(0, 2 * np.pi, num_points)
    costheta = np.random.uniform(-1, 1, num_points)
    theta = np.arccos(costheta)
    x = np.sin(theta) * np.cos(phi)
    y = np.sin(theta) * np.sin(phi)
    z = np.cos(theta)
    return np.stack((x, y, z), axis=-1)

def plot_electric_field(point_charges, observation_points, title):
    X, Y = np.meshgrid(observation_points, observation_points)
    Ex, Ey = np.zeros_like(X), np.zeros_like(Y)

    for charge in point_charges:
        for i in range(len(observation_points)):
            for j in range(len(observation_points)):
                field = charge.computeField((X[i, j], Y[i, j]))
                Ex[i, j] += field[0]
                Ey[i, j] += field[1]

    magnitude = np.sqrt(Ex**2 + Ey**2)
    magnitude[magnitude == 0] = 1  # Avoid division by zero
    Ex /= magnitude
    Ey /= magnitude

    plt.quiver(X, Y, Ex, Ey, scale=50)
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title(title)
    plt.show()

if __name__ == "__main__":
    # Define parameters
    num_points = 100
    q = 1
    Q = 100 * q
    r = 1

    # Generate surface points on the sphere
    surface_points = generate_surface_points(num_points)

    # Create pointCharge objects for each surface point
    point_charges = [pointCharge(q, point * r) for point in surface_points]

    # Define observation points for plotting the field in the xy-plane
    observation_points = np.linspace(-2, 2, 50)

    # Plot electric field produced by 100 point charges on the sphere's surface
    plot_electric_field(point_charges, observation_points, "Electric Field by 100 Charges on Sphere's Surface")
````
